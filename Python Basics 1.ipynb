## Operators

```python
# floating point division /
# integer division //

print(10 + 5)
print(10-5)
print(10 * 5)
print(10 / 5)
print(10 // 3)
print(10 % 3)
print(10 ** 3)
```

```python
# Relational Operator - Returns Boolean value
print(2<4)
print(2==4)
print(4!=5)
```

```python
# Logical Operator
print(2<4 and 2<1)
print(2<4 or 2<1)
print(not 1)
```

```python
# left shift operator: a X 2^b
a=2
b=3
a << b
```

```python
# right shift operator a // 2^b
a=10
b=2
a >> b
```

## Conditional Statement

```python
age= 25
income =50000

condition=age>10 and income >30000
print(condition)
```

```python
# Basic if statement
if condition == True:
    print("you are eligible")
```

```python
if age>18 and income>30000:
    print("you are eligible")
```

```python
# If else
money=5
if money>=10:
    print("buy Lunch")
else:
    print("Bring lunch")
```

```python
# If  Elif else
grade=85
if grade>=90:
    print("A")
elif grade>=80:
    print("B")
else:
    print("C")
```

```python
 # Nested Ifs
x=10
if x>5:
     print("above 5")
     if x>8:
         print("Also above 8")
```
