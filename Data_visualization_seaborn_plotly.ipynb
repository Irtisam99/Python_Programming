```python

import seaborn as sns
print(sns.__version__)
```

```python
sns.set_theme(
    style='whitegrid',
    palette='flare',
    font='Times New Roman',
    font_scale=1

)
```

```python
"""
Statistical plot: Historical plot
1. Relational Plot (relplot): establishes relation between variables
    X=1 ; Y=2
    . Independent variables
    . Dependent variables
2. Distributional plot(displot) :  visualizes the distribution of the variables
3.Categorical plot (catplot) : provides statistical information about the categorical variables

matplotlib > seaborn > plotly ( interactive)
"""



sns.get_dataset_names()
```

```python
tips=sns.load_dataset('tips')
fmri=sns.load_dataset('fmri')
penguins=sns.load_dataset('penguins')
```

```python
tips.head()
```

```python
"""
Relational Plots
"""

# Scatter Plot
plot=sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    kind='scatter'
)

__=plot.set(
    xlabel='Total bills(USD)',
    ylabel='Tips(USD)',
    title='Relation between Total Bill and Tip (USD) '
)
```

```python
tips['day'].value_counts()
```

```python
plot=sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    kind='scatter',
    hue='day',
    row='day',
    height=4,

)
```

```python
# Line plot

plot=sns.relplot(
    data=fmri,
    x='timepoint',
    y='signal',
    kind='line',
    height=4,
    col='region',
    hue='region'
)
plot.set(
    xlabel='Timeplots',
    ylabel='Signals',
)
```

```python
"""
Distributional Plots
1. Univariate plot ( one variable)
2. Bivariate plot ( two variables)
3. Mutlivariate plot ( three or more variables/features)
"""
penguins.head()
```

```python
# By default histogram , bins=10
sns.displot(
    data=penguins,
    x='flipper_length_mm',
    kind='hist',
    bins=20,

)
```

```python
sns.displot(
    data=penguins,
    x='species',
    discrete=True,
    shrink=0.8,


)
```

```python
sns.displot(
    data=penguins,
    x='flipper_length_mm',
    hue='species',
    col='species',
)
```

```python
sns.displot(
    data=penguins,
    x='flipper_length_mm',
    hue='species',
    element='step',
    multiple='stack',
    stat='probability',
)
```

```python
# Kernel Density Estimation ( bell curve)

sns.displot(
    data=penguins,
    x='flipper_length_mm',
    kind='kde',
    fill=True,
    hue='species',
    multiple='stack',

)
```

```python
# Bivariate Distribution : Heatmap
sns.displot(
    data=penguins,
    x='bill_length_mm',
    y='bill_depth_mm',
    cbar=True
)
```

```python
sns.pairplot(penguins)
```

```python
"""
Categorical Plot

Strip Plot
- It shows individual data points as a scatter plot along cat axis
- It visualizes outliers, clusters and distribution
- Limitations: data can overlap , making it hard to understand the distribution
"""
sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    kind='strip'

)
```

```python
sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    kind='strip',
    jitter=False,
    order=['Sun','Sat','Thur','Fri']

)
```

```python
"""
Swarm plot
- it is similar to strip plot but adjust the points to avoid overlap
- it can visualize outliers , clusters and distribution for small dataset
- limitation: Doesnt not fit well for larger datasets for spacing issues

"""

sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    kind='swarm',
)
```

```python
"""
Box plot
- Statistics:
min , Q1,median,Q3,Max
- It shows distribution through statistics
- it visualizes outliers , distributions
-Limitation: No info present for understanding data density
"""

sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    kind='box',
)
```

```python
"""
Violin Plot
- It combines box plot and distribution density graph ( KDE)
- it visualizes distribution and statistics
"""

sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    kind='violin'

)
```

```python
"""
Data Distribution with plotly

Scatter plot with plotly

sns                px
kind=scatter      px.scatter
data              data_frame
hue                color
col                facet_col
row                facet_row
"""
```

```python
import plotly.express as px

import plotly.io as pio
plot=px.scatter(
    data_frame=tips,
    x='total_bill',
    y='tip',
    color='sex',
    facet_col='time',
    symbol='smoker'
)
plot.update_layout(
    xaxis_title='Total Bill',
    yaxis_title='Tips',
    title='Total bill vs Tips',
)
plot.write_html("plot.html")  # Navigate to plot.html file
plot.show()

```
