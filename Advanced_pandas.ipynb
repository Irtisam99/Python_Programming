```python
"""
 Data Sanitization/Cleaning

 Challenge: Contains missing values

 Approach 1: Removing the rows contains missing values
 Approach 2: Impute the missing values with statistical information/ contextual information

"""
df.isnull().sum()
```

```python
mask=df.isnull().any(axis=1)
df[mask]
```

```python
avg_calorie=df['Calories'].mean()
```

```python
df['Calories']=df['Calories'].fillna(avg_calorie)
df[mask]
```

```python
# For dropping
mask=df.isnull().any(axis=1)
~mask
```

```python
df=df[~mask]
df.isnull().sum()
```

```python
"""
Remove Duplicates
"""
mask=df.duplicated()
df[mask]
```

```python
df=df[~mask]
df.duplicated().sum()
```

```python
# Another method : Drop the duplicate rows using drop_duplicates()
# implace=true : to make changes on the main data frame rather than copying
df=df.drop_duplicates(inplace=True)

```
