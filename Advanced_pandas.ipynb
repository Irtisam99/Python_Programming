```python
"""
Remove Outliers
IQR - Interquartile Range

COls=[1,4,2,3,1,2,100]
L,R = Expected Data values lies in the range
<L, R> == Outliers

IQR=Q3-Q1 (75%-25%)
L=Q1-1.5*IQR
R=Q3+1.5*IQR

Any values outside the L and R are outliers
"""

df.describe()
```

```python
Q1=df['Pulse'].quantile(0.25)
Q3=df['Pulse'].quantile(0.75)
IQR=Q3-Q1
print(IQR)
L=Q1-1.5*IQR
R=Q3+1.5*IQR
print(L,R)
```

```python
mask_outlier=df['Pulse'].between(L,R)
df[~mask_outlier]
```

```python
df=df[mask_outlier]
```

```python
"""
Col A,B,C (Target)

Correlation displays the relation between two variables
-Positive Correlation: if x increases, y increases and vice versa
-Negative Correlation: if x increases,y decreases and vice versa

Strength of correlation: [-1.1]
col A and col B correlation value is 1  ( Strongly positively correlated )
col A and col B correlation value is -1  ( Strongly negatively correlated )
col A and col B correlation value is 0  (No correlation)

"""
```

```python
"""
Predictive analysis
- Normalizing ( Numerical Columns)
"""
def normalize(col):
    mean=col.mean()
    std=col.std()
    return (col-mean)/std

df['Pulse']=df['Pulse'].agg(normalize)
df['Maxpulse']=df['Maxpulse'].agg(normalize)
df['Duration']=df['Duration'].agg(normalize)
df['Calories']=df['Calories'].agg(normalize)


```
