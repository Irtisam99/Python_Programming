```python
import seaborn as sns
print(sns.__version__)
```

```python
sns.set_theme(
    style='whitegrid',
    palette='flare',
    font='Times New Roman',
    font_scale=1

)
```

```python
"""
Statistical plot: Historical plot
1. Relational Plot (relplot): establishes relation between variables
    X=1 ; Y=2
    . Independent variables
    . Dependent variables
2. Distributional plot(displot) :  visualizes the distribution of the variables
3.Categorical plot (catplot) : provides statistical information about the categorical variables

matplotlib > seaborn > plotly ( interactive)
"""



sns.get_dataset_names()
```

```python
tips=sns.load_dataset('tips')
fmri=sns.load_dataset('fmri')
penguins=sns.load_dataset('penguins')
```

```python
tips.head()
```

```python
"""
Relational Plots
"""

# Scatter Plot
plot=sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    kind='scatter'
)

__=plot.set(
    xlabel='Total bills(USD)',
    ylabel='Tips(USD)',
    title='Relation between Total Bill and Tip (USD) '
)
```

```python
tips['day'].value_counts()
```

```python
plot=sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    kind='scatter',
    hue='day',
    row='day',
    height=4,

)
```

```python
# Line plot

plot=sns.relplot(
    data=fmri,
    x='timepoint',
    y='signal',
    kind='line',
    height=4,
    col='region',
    hue='region'
)
plot.set(
    xlabel='Timeplots',
    ylabel='Signals',
)
```

```python
"""
Distributional Plots
1. Univariate plot ( one variable)
2. Bivariate plot ( two variables)
3. Mutlivariate plot ( three or more variables/features)
"""
penguins.head()
```

```python
# By default histogram , bins=10
sns.displot(
    data=penguins,
    x='flipper_length_mm',
    kind='hist',
    bins=20,

)
```

```python
sns.displot(
    data=penguins,
    x='species',
    discrete=True,
    shrink=0.8,


)
```

```python
sns.displot(
    data=penguins,
    x='flipper_length_mm',
    hue='species',
    col='species',
)
```

```python
sns.displot(
    data=penguins,
    x='flipper_length_mm',
    hue='species',
    element='step',
    multiple='stack',
    stat='probability',
)
```

```python
# Kernel Density Estimation ( bell curve)

sns.displot(
    data=penguins,
    x='flipper_length_mm',
    kind='kde',
    fill=True,
    hue='species',
    multiple='stack',

)
```

```python
# Bivariate Distribution : Heatmap
sns.displot(
    data=penguins,
    x='bill_length_mm',
    y='bill_depth_mm',
    cbar=True
)
```
